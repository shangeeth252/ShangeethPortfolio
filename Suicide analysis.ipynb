__We will now check the amount of data distributed across different groups. First we can see that in all the age groups there is about an equal amount of data.__

```python
p=sns.countplot(x='age',data=dataframe)
p.set_xticklabels(p.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

__The same goes for both genders as the data is distributed equally.__

```python
p=sns.countplot(x='sex',data=dataframe)
```

__Now that we confirmed that there is a fair amount of data across various categories, we will now analyze what gender has more suicides identified by age category.__

```python
a = sns.barplot(x='sex', y='suicides_no', hue='age', data=dataframe)
```

__By a very surprising amount it looks like males have higher values than females in suicides. We can also see that the age 35-54 has predominantly high amounts of suicide in both genders.__

__We will also analyze the same data but by generation to get a better idea of which generation has high suicide values.__

```python
g = sns.barplot(x='sex', y='suicides_no', hue='generation', data=dataframe)
```

__According to the data, it is evident that the "Boomer" generation has very high suicide rates followed by the silent generation on both genders. The Generation with the least is generation Z as they are all kids during this time period.__

__Next we will see the values of the suicides across the various years in different graphs. It will be easier to visualize the suicide values to see which years have high suicide rates.__

```python
p = sns.catplot('sex','suicides_no',hue='age',col='year',data=dataframe,kind='bar',col_wrap=5)
```

__Now we will visualize the change in suicide over the years with line graphs. This can be a way to see the change in suicides because of various events that occurs over the years.__  

```python
 age_5 = dataframe.loc[dataframe.loc[:, 'age']=='5-14 years',:]
age_15 = dataframe.loc[dataframe.loc[:, 'age']=='15-24 years',:]
age_35 = dataframe.loc[dataframe.loc[:, 'age']=='35-54 years',:]
age_75 = dataframe.loc[dataframe.loc[:, 'age']=='75+ years',:]
age_25 = dataframe.loc[dataframe.loc[:, 'age']=='25-34 years',:]
age_55 = dataframe.loc[dataframe.loc[:, 'age']=='55-74 years',:]
```

```python
j = sns.lineplot(x='year', y='suicides_no', data=age_5)
k = sns.lineplot(x='year', y='suicides_no', data=age_15)
l = sns.lineplot(x='year', y='suicides_no', data=age_25)
m = sns.lineplot(x='year', y='suicides_no', data=age_35)
n = sns.lineplot(x='year', y='suicides_no', data=age_55)
o = sns.lineplot(x='year', y='suicides_no', data=age_75)

f = plt.legend(['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'])
```

__As you can see, there is a steady trend over the years 1990-2015. However, the suicide numbers are still fairly high which is not a good thing according to the age group 35-54.__ 

__Next we will analyze the line graph using the male and female population.__

```python
male_population = dataframe.loc[dataframe.loc[:, 'sex']=='male',:]
female_population = dataframe.loc[dataframe.loc[:, 'sex']=='female',:]
```

```python
p = sns.lineplot(x='year', y='suicides_no', data=male_population)
q = sns.lineplot(x='year', y='suicides_no', data=female_population)

_ = plt.legend(['males', 'females'])
```

__As expected, the male values are very high compared to the female values. The trends on both seem to be very steady on both populations and doesn't seem to change much over the years.__

__In conclusion, we have witnessed that males tend to have higher suicide values compared to females regardless of year and country.
